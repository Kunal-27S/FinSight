```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv('../../data/stock_prices/apple_cleaned.csv')
data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data['date'] = pd.to_datetime(data['date'])
plt.figure(figsize=(10,5))
plt.plot(data['date'], data['close'], label='Close Price')
plt.title('Apple Stock Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```

```python
data['returns'] = data['close'].pct_change()
```

```python
plt.figure(figsize=(10,5))
plt.plot(data['date'], data['returns'])
plt.title('Daily Returns')
plt.xlabel('Date')
plt.ylabel('Return')
plt.show()
```

```python
data['ma_7'] = data['close'].rolling(7).mean()
data['ma_30'] = data['close'].rolling(30).mean()

plt.figure(figsize=(10,5))
plt.plot(data['date'], data['close'], label='Close', alpha=0.6)
plt.plot(data['date'], data['ma_7'], label='7-Day MA')
plt.plot(data['date'], data['ma_30'], label='30-Day MA')
plt.legend()
plt.show()
```
